```python
import pandas as pd
import numpy as np
url ="https://raw.githubusercontent.com/ziafaq/genai/refs/heads/main/diabetes.csv"
df =pd.read_csv(url)
df.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense,Input

def create_model():
  # create the model
  model = Sequential()

  # add input layer
  #model.add(Input(shape=(8,)))
  model.add(Input(shape=(df.shape[1]-1,)))

  # add hidden layer
  model.add(Dense(2, activation='relu'))

  # output layer
  # for classification use 'sigmoid' as the activation function 0 1
  # for regression use 'linear' (default)
  model.add(Dense(1, activation='sigmoid'))

  # regression = 'mse'
  # classification 'binary loss' # binary_crossentopy (loss/diff b/w actual/predicted)
  model.compile(optimizer='adam', metrics=['accuracy'], loss='binary_crossentropy')

  return model
```

```python
# train test split

display(df)

X = df.iloc[:,0:8]

Y = df.iloc[:,8]

!pip install scikit-learn==1.4.2
```

```python
!pip install scikeras

from scikeras.wrappers import KerasClassifier

# # create the model
model = KerasClassifier(model=create_model)

batch_size = [10]
epochs = [1]
param_grid = dict(batch_size=batch_size, epochs=epochs)
```

```python
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(estimator=model,param_grid=param_grid)

grid_result = grid.fit(X,Y)
```

```python
from keras.layers import Dropout
# create model in a function
def create_model3(act_fn, neurons, initalizr):
  model = Sequential()
  model.add(Input(shape=(8,)))
  model.add(Dense(neurons, kernel_initializer=initalizr, activation=act_fn))
  model.add(Dropout(0.2))
  model.add(Dense(1, activation='sigmoid'))
  #model.compile(optimizer='Adam', metrics=['accuracy'], loss='binary_crossentropy')

  return model
```

```python

# 7 hyperparameters and two values each
# batch_size , epochs, optimizer, learning_rate, act_fn

batch_size = [100]
epochs = [1]
optimizer = ['SGD'] #, 'RMSprop']
learn_rate = [0.001]
initilzr = ['uniform']
act_fn = ['tanh']
loss = ['binary_crossentropy']
n_neurons = [10]

model = KerasClassifier(model=create_model3)
param_grid = dict(batch_size=batch_size, epochs=epochs, optimizer=optimizer, loss=loss, model__initalizr=initilzr,
                  optimizer__learning_rate=learn_rate, model__act_fn=act_fn, model__neurons=n_neurons)

grid = GridSearchCV(estimator=model,param_grid=param_grid) #model = KerasClassifier(model=create_model)
grid_result = grid.fit(X,Y)
print(grid_result.best_score_, grid_result.best_params_)
```
